## LLMs what are they good for?

Large language models (LLMs) are powerful AI systems that can generate natural language texts based on a given input. They’ve been used for a number of tasks in the field of natural language processing (NLP)[^2]. These have included text summarization, translation, sentiment analysis (e.g. tell you if your text has a positive or negative sentiment), etc.

[^2]: NLP -Natural Language Processing

One of the most exciting applications of LLMs has been the creation of interactive question and answer chatbots (these are AI models that you can prompt with questions). These have been used to augment writing, have conversations with, help generate code and more! If you’ve tried [OpenAI’s ChatGPT](https://openai.com/blog/chatgpt) then you know how powerful a tool it can be.

## LLMs how do they work?

LLMs work through a number of different complex steps (greatly oversimplified):

-   Words are first converted into high dimensional numeric vectors, representing their semantic meanings. Think of techniques like [Word2Vec](https://en.wikipedia.org/wiki/Word2vec) or [GloVe](https://en.wikipedia.org/wiki/GloVe)[^3]

-   These vectors are fed into a massive machine learning model (typically a [Transformer architecture](https://en.wikipedia.org/wiki/Transformer_(deep_learning_architecture)%20with%20attention%20mechanisms) that learns patterns, probabilities, and relationships between words.

-   This training enables the model to generate coherent and contextually relevant text based on new input it receives.

[^3]: GloVe - Global Vectors for Word Representation

## What is Gemma?
<img src="gemma_logo.png" alt="image" width="45%" height="auto" style="border: none;">
<!-- ![](gemma_logo.png) -->

[Gemma](https://blog.google/technology/developers/gemma-open-models/) is described as a family of lightweight models that was developed by Deepmind and other teams at Google. These models were used to create Google’s Gemini models. The name of the model comes from the latin word gemma, translating to gem or precious stone.

The open source [Gemma model](https://storage.googleapis.com/deepmind-media/gemma/gemma-report.pdf) is based on the transformer decoder architecture and comes in two sizes: a 7B and 2B. These represent the number of parameters that each model has, which are actually 7,751,248,896 (7B) and 1,981,884,416 (2B). On benchmarks the Gemma model performs on par or better than the LLaMA and Mistral models. You can check out the interactive benchmark comparisons [here](https://ai.google.dev/gemma#research).



## What is Mistral?
<img src="mistral_logo.svg" alt="image" width="45%" height="auto" style="border: none;">


[Mistral](https://mistral.ai/news/announcing-mistral-7b/) was created by the Paris-based Mistral AI team. The founders of the team came from Meta Platforms and Google's DeepMind. Similar to the Gemma model the Mistral model (specifically the Mistral-7B model) is based on the transformer architecture and has about 7 billion parameters. The name [*Mistral*](https://en.wikipedia.org/wiki/Mistral_(wind)) presumably comes from the the term describing strong, cold, winds that blow through France from south to the Mediterranean.

## What is Zephyr?
<img src="zephyr_logo.png" alt="image" width="45%" height="auto" style="border: none;">

The Zephyr models are a series of models that were fine tuned by the team at [Hugging Face](https://huggingface.co/HuggingFaceH4). The [Zephyr 7B-β](https://huggingface.co/HuggingFaceH4/zephyr-7b-beta) model is the second of  (currently) 3 models and is a fine tuned version of Mistral 7B. The idea behind model fine-tuning is to improve the accuracy of the models. In this case the Mistral model was additionally fine-tuned with [AI feedback (AIF)](https://arxiv.org/abs/2310.16944). Similar to the terms *mistral* the word [*zephyr*](https://www.oed.com/search/dictionary/?scope=Entries&q=zephyr) refers to a soft gentle breeze. 

One of the latest Zephyr models,  [Zephyr 7B Gemma](https://huggingface.co/HuggingFaceH4/zephyr-7b-gemma-v0.1) is a fine-tuned version of the Gemma model by Google, however the has yet been configured for inference on Hugging Face. There is an interactive demo availble though, check it out [here](https://huggingface.co/spaces/HuggingFaceH4/zephyr-7b-gemma-chat).
